```python
import numpy as np
import pandas as pd
from pandas.util.testing import makeTimeSeries

s = makeTimeSeries(5)

cols = ['col_1', 'col_2']
df = pd.DataFrame(abs(np.random.randn(5, 2)), columns=cols)
df['date'] = s.index

from datetime import datetime

df['days since'] = (df['date'] - datetime.today()).dt.days
# unformatted df
display(df)
```

```python
# styled df. note conventional methods no longer work
df_styled = df.style.background_gradient(cmap='Greens')
df_styled
```
