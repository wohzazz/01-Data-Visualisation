## export images to word document

```python
# !pip install docx

# ! pip install python-docx # docx doesn't work. see https://stackoverflow.com/questions/22765313/when-import-docx-in-python3-3-i-have-error-importerror-no-module-named-excepti
```

```python
def export_to_word(dpi =72):

    import docx
    from docx.shared import Inches, RGBColor, Pt
    from docx.enum.style import WD_STYLE_TYPE


    # create instance of word document
    doc = docx.Document()

    # specify export width and height
    ex_width = Inches(6)
    ex_height = Inches(4)

    # define styles
    style = doc.styles['Normal']
    font = style.font
    font.name = 'Calibri'
    font.size = Pt(10)
    # font.italic = False

    # define styles for other elements
    font_styles = doc.styles
    font_charstyle = font_styles.add_style('CaptionStyle', WD_STYLE_TYPE.PARAGRAPH)
    font_object = font_charstyle.font
    font_object.size = Pt(8)
    font_object.name = 'Calibri'
    font_object.color.rgb = RGBColor(68,114,196)
    font_object.italic = True

    # add headings
    _ = doc.add_heading("Exported Seaborn Visuals", 0)

    if dpi ==72:
        image = r'images\export_example_1.png'
    else:
        image = r'images\export_example_1_{dpi}dpi.png'.format(dpi=dpi)
    _ = doc.add_picture(image, width = ex_width, height = ex_height)

    if dpi ==72:
        image = r'images\export_example_2.png'
    else:
        image = r'images\export_example_2_{dpi}dpi.png'.format(dpi=dpi)
    _ = doc.add_picture(image, width = ex_width, height = ex_height)

    # export file
    output_file = f'Export Seaborn Visual Test {dpi}dpi.docx'
    doc.save(output_file)

    # open word document for review
    import os
    os.startfile(output_file)
    

export_to_word()
export_to_word(dpi=300)
export_to_word(dpi=600)
```

Overall thoughts
- Note there is not a great discernible difference between 300dpi and 600dpi but 72dpi definitely appears low res
- 600dpi is almost double the size of 300dpi
- Recommend using 300dpi for most applications

### multiple plots, column by 1 dimension using a specific order

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.relplot(x="total_bill"
                ,y="tip"
                ,data=tips, kind = "scatter", col = 'sex', col_order = ['Female', 'Male'])
plt.show()
```

# Bar Plots

- Displays mean of quantitative variable per category
- Lines show 95% confidence intervals for the means!
- Show uncertainty about our estimate
- Assume our data is a random sample

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')


sns.catplot(x="day", y = "total_bill", data = tips, kind = "bar", order = ["Sun", "Thur", "Fri", "Sat"])
            
plt.show()
```

### whiskers can also change
- By default it extends to 1.5 * interquartile range (difference between 75th and 25th percentile) (TBC- why is the 1.5*interquartile range not symmetrical?)
- Specifying whis = 2 changes this to 2 * interquartile range
- Specifying whis = [5,95] shows 5th and 95th percentile respectively

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.catplot(x="time", y = "total_bill", data = tips, kind = "box", whis = [5,95])
            
plt.show()
```

# point plot

Great for comparing means and CI of means across different groups and subgroups

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.catplot(x="smoker", y = "total_bill", data = tips, kind = "point")
            
plt.show()
```

## remove the join lines

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.catplot(x="smoker", y = "total_bill", data = tips, kind = "point", join =False)
            
plt.show()
```

## can be used to compare medians

Median are more robust to outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median

tips = sns.load_dataset('tips')

sns.catplot(x="smoker", y = "total_bill", data = tips, kind = "point", estimator = median)
            
plt.show()
```

## set capsize of CI

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.catplot(x="smoker", y = "total_bill", data = tips, kind = "point", capsize = 0.2)
            
plt.show()
```
